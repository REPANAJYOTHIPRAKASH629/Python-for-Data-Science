```python
import numpy as np
import pandas as pd
import statistics as st
```

```python
#Standard Deviation
a=np.array([5,6,7])
print(a)
print(np.std(a))
```

```python
#Varience
a=np.array([[1,2],[3,4]])
print(a)
print(np.var(a))
```

```python
#Median
a=np.array([1,2,3])
print(a)
print(np.median(a))
```

```python
a=np.array([1,2,3,3,3,4])
print(a)
print(st.mode(a))
```

```python
#Mean 
a=np.array([5,6,7])
print(a)
b=np.mean(a)
print(b)
```
